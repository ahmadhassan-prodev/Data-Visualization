```python
import matplotlib.pyplot as plt
Class = [9,10,11,12]
marks = [478,1044,493,1022]

plt.plot(Class, marks)
plt.xlabel("Class")
plt.ylabel("Marks Obtained")
plt.title("Ahmad's Marks in class")
```

```python
ahmad_marks = [478,1044,493,1022]
awais_marks = [490,1058,470,980]

plt.plot(Class, ahmad_marks, label="Ahmad Hassan")
plt.plot(Class, awais_marks, label="Awais Ali")
plt.xlabel("Class")
plt.ylabel("Marks")
plt.title("Marks Comaprison")
plt.legend()
plt.show()
```

```python
ahmad_marks = [478,1044,493,1022]
awais_marks = [490,1058,470,980]

plt.plot(Class, ahmad_marks, "g^:" ,label="Ahmad Hassan")
plt.plot(Class, awais_marks, "ro--" ,label="Awais Ali")
plt.xlabel("Class")
plt.ylabel("Marks")
plt.title("Marks Comaprison")
plt.legend()
plt.show()
```

```python
ahmad_marks = [478,1044,493,1022]
awais_marks = [490,1058,470,980]

plt.plot(Class, ahmad_marks, color="green", linestyle="--", label="Ahmad Hassan")
plt.plot(Class, awais_marks, color="yellow", linestyle="-", label="Awais Ali")
plt.xlabel("Class")
plt.ylabel("Marks")
plt.title("Marks Comaprison")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
